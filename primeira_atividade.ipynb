```python
import sqlite3

conn = sqlite3.connect("database.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS countries (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    code TEXT UNIQUE
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS country_data (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    country_id INTEGER,
    date TEXT,
    population INTEGER,
    gdp INTEGER,
    FOREIGN KEY(country_id) REFERENCES countries(id)
)
""")

conn.commit()
```
